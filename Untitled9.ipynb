```python
import os
import json
import cv2
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import models, ops
from torchvision.transforms import ToTensor, Normalize, Compose
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Charger les annotations JSON
def load_annotations(json_path):
    with open(json_path, 'r') as f:
        annotations = json.load(f)
    return annotations

# Dataset pour la dtection d'objets
class ObjectDetectionDataset(Dataset):
    def __init__(self, annotations, images_dir, transform=None, label_map=None):
        self.annotations = annotations
        self.images_dir = images_dir
        self.transform = transform
        self.label_map = label_map or self.build_label_map()

    def build_label_map(self):
        labels = set()
        for annotation in self.annotations:
            for obj in annotation["annotations"]:
                labels.add(obj["label"])
        return {label: idx for idx, label in enumerate(sorted(labels))}

    def __len__(self):
        return len(self.annotations)

    def __getitem__(self, idx):
        annotation = self.annotations[idx]
        image_name = annotation["image"]
        image_path = os.path.join(self.images_dir, image_name)

        # Charger l'image
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Charger les bounding boxes et labels
        boxes = []
        labels = []
        for obj in annotation["annotations"]:
            x, y, w, h = obj["coordinates"].values()
            x_min = x - w / 2
            y_min = y - h / 2
            x_max = x + w / 2
            y_max = y + h / 2
            boxes.append([x_min, y_min, x_max, y_max])
            labels.append(self.label_map[obj["label"]])

        boxes = torch.tensor(boxes, dtype=torch.float32)
        labels = torch.tensor(labels, dtype=torch.int64)

        # Appliquer les transformations
        if self.transform:
            image = self.transform(image)

        return image, {"boxes": boxes, "labels": labels}

# Modle Fast R-CNN
class FastRCNN(nn.Module):
    def __init__(self, num_classes, roi_size=(7, 7), hidden_dim=1024):
        super(FastRCNN, self).__init__()
        backbone = models.resnet50(pretrained=True)
        self.feature_extractor = nn.Sequential(*list(backbone.children())[:-2])
        self.roi_pool = ops.roi_pool
        self.fc1 = nn.Linear(2048 * roi_size[0] * roi_size[1], hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        self.cls_head = nn.Linear(hidden_dim, num_classes + 1)  # +1 pour le fond
        self.reg_head = nn.Linear(hidden_dim, num_classes * 4)  # 4 coordonnes par classe

    def forward(self, images, proposals, gt_boxes=None, gt_labels=None):
        feature_map = self.feature_extractor(images)
        # ROI Pooling
        roi_out = self.roi_pool(feature_map, proposals, output_size=(7, 7))
        roi_out = roi_out.flatten(start_dim=1)

        # Fully Connected Layers
        out = F.relu(self.fc1(roi_out))
        out = F.relu(self.fc2(out))

        # Classification et Rgression
        class_logits = self.cls_head(out)
        bbox_preds = self.reg_head(out)

        return class_logits, bbox_preds

# Fonction de post-traitement
def postprocess_predictions(predictions, proposals, iou_thresh=0.5):
    class_logits, bbox_preds = predictions
    scores, labels = torch.max(F.softmax(class_logits, dim=1), dim=1)
    keep_indices = ops.nms(proposals, scores, iou_thresh)
    return proposals[keep_indices], scores[keep_indices], labels[keep_indices]

# Visualiser les prdictions
def visualize_predictions(image, proposals, scores, labels, label_map):
    label_map_rev = {v: k for k, v in label_map.items()}
    fig, ax = plt.subplots(1, figsize=(12, 8))
    ax.imshow(image.permute(1, 2, 0).cpu().numpy())
    for i, box in enumerate(proposals):
        x_min, y_min, x_max, y_max = box.cpu().numpy()
        label = label_map_rev[labels[i].item()]
        score = scores[i].item()
        rect = patches.Rectangle((x_min, y_min), x_max - x_min, y_max - y_min,
                                 linewidth=2, edgecolor="r", facecolor="none")
        ax.add_patch(rect)
        ax.text(x_min, y_min - 5, f"{label}: {score:.2f}", color="white",
                fontsize=12, backgroundcolor="red")
    plt.show()

# Main
if __name__ == "__main__":
    annotations_file = "/content/drive/MyDrive/Colab Notebooks/Dataset/train/_annotations.createml.json"
    images_dir = "/content/drive/MyDrive/Colab Notebooks/Dataset/train"
    annotations = load_annotations(annotations_file)

    transform = Compose([
        ToTensor(),
        Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    dataset = ObjectDetectionDataset(annotations, images_dir, transform=transform)
    dataloader = DataLoader(dataset, batch_size=1, shuffle=True, collate_fn=lambda x: tuple(zip(*x)))

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    num_classes = len(dataset.label_map)  # Exemple : plastique, carton, verre, mtal
    model = FastRCNN(num_classes).to(device)
    model.eval()  # Pour l'infrence

    # Charger un exemple
    for images, targets in dataloader:
        images = torch.stack(images).to(device)
        with torch.no_grad():
            # Simuler des propositions pour l'exemple
            gt_boxes = targets[0]["boxes"]
            nms_indices = ops.nms(gt_boxes, torch.ones(len(gt_boxes)), 0.5)
            proposals = gt_boxes[nms_indices]  # GT comme propositions
            predictions = model(images, [proposals.to(device)])

            # Post-traitement
            proposals, scores, labels = postprocess_predictions(predictions, proposals)

            # Visualisation
            visualize_predictions(images[0], proposals, scores, labels, dataset.label_map)
        break
```
